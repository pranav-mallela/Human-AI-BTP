```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
accuracies = [[0.35, 0.65, 0.70, 0.56, 0.66],
              [0.89, 0.47, 0.41, 0.45, 0.62, 0.37, 0.45],
              [0.70, 0.29, 0.36, 0.92, 0.54, 0.55, 0.52, 0.44, 0.68, 0.61],
              [0.53, 0.36, 0.50, 0.32, 0.35, 0.56, 0.52, 0.86, 0.72, 0.31, 0.69, 0.85, 0.36],
              [0.48, 0.94, 0.53, 0.70, 0.43, 0.58, 0.30, 0.88, 0.62 , 0.57, 0.38, 0.49, 0.39, 0.55, 0.29]]

test = [10,100,999,9999]

B = [5,10,15]
```

```python
i = 0
j = 0
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))
for a in accuracies:
    selectall = []
    random = []
    pseudolb = []
    eamc5 = []
    eamc10 = []
    eamc15 = []
    pomc = []
    a.sort()
    a.reverse()
    for t in test:
        for b in B:
            csv_file_path = f'./output/eamc5/cnn_data_accuracy_{a}_{t}_0.csv'
            with open(csv_file_path, 'r') as file:
                csv_reader = csv.reader(file)
                next(csv_reader)
                for row in csv_reader:
                    if b==5:
                        selectall.append(100-(float(row[5])*100))
                        random.append(100-(float(row[6])*100))
                        pseudolb.append(100-(float(row[8])*100))
                        pomc.append(100-(float(row[9])*100))
                        eamc5.append(100-(float(row[10])*100))
                    # elif b==10:
                    #     eamc10.append(100-(float(row[2])*100))
                    # else:
                    #     eamc15.append(100-(float(row[2])*100))

    x = ['1e1','1e2','1e3','1e4'] 
    print(a)
    if(len(a)==5):
        i=0
        j=0
    if(len(a)==7):
        i=0
        j=1
    if(len(a)==10):
        i=0
        j=2
    if(len(a)==13):
        i=1
        j=0
    if(len(a)==15):
        i=1
        j=1
    
    axes[i,j].plot(x, selectall, label='Select All', marker='o',color='cyan')
    axes[i,j].plot(x, random, label='Random', marker='s',color='green')
    axes[i,j].plot(x, eamc5, label='EAMC (B=5)', marker='*',color='blue')
    axes[i,j].plot(x, pseudolb, label='Pseudo LB', marker='^',color='red')
    axes[i,j].plot(x, pomc, label='POMC', marker='.',color='orange')

for row in axes:
    for ax in row:
        ax.set_xlabel('Number of Training Instances')
        ax.set_ylabel('Error Rate(%)')

axes[1,2].axis('off')
axes[1,2].plot([],[], label='Select All', marker='o',color='cyan')
axes[1,2].plot([], [], label='Random', marker='s',color='green')
axes[1,2].plot([], [], label='EAMC (B=5)', marker='*',color='blue')
axes[1,2].plot([], [], label='Pseudo LB', marker='^',color='red')
axes[1,2].plot([], [], label='POMC', marker='.',color='orange')
axes[1,2].legend(loc='center')
plt.tight_layout()
plt.savefig('customcnn_CIFAR.png',bbox_inches='tight')
plt.show()
    
```
